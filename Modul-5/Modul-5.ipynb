```python
import numpy as np #untuk memanggil library numpy dialiaskan sebagai np
import cv2  #untuk memanggil library cv2
import matplotlib.image as im  #untuk memanggil library matplotlib
import matplotlib.pyplot as plt #untuk memanggil library matplotlib.pyplot dialiaskan sebagai plt
#Import Library
```

```python
image=im.imread("rem_nangis.jpg")
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
```

```python
def mean (img,h,w):
    panjang,lebar=img.shape
    padH,padW= int ((h-1)/2),int ((w-1)/2)
    result=np.zeros(img.shape,dtype=int)
    imgPad=np.pad(img,((padH,h-padH-1),(padW,w-padW-1)),'constant')
    for i in range (panjang):
        for j in range (lebar):
            result[i,j]=np.mean(imgPad[i:i+h,j:j+w])
    return result
```

Kode diatas merupakan kode yag digunakan untuk menentukan mean filtering. Langkah yang dilakukan yaitu dengan memasukkan nilai matriksnya terlebih dahulu kemudian menghitung matriksnya sesuai rumus mean. Hasil yang diperoleh di olah lagi dengan cara melakukan perulangan pada tinggi dan lebar dari image

```python
    def median (img,h,w):
        panjang,lebar=img.shape
        padH,padW= int ((h-1)/2),int ((w-1)/2)
        result=np.zeros(img.shape,dtype=int)
        imgPad=np.pad(img,((padH,h-padH-1),(padW,w-padW-1)),'edge')
        for i in range (panjang):
            for j in range (lebar):
                result[i,j]=np.median(imgPad[i:i+h,j:j+w])
        return result
```

Kode diatas merupakan kode yang digunakan untuk menentukan median filtering. Langkah yang dilakukan yaitu dengan memasukkan nilai matriksnya terlebih dahulu kemudian menghitung matriksnya sesuai rumus median. Hasil yang diperoleh di olah lagi dengan cara melakukan perulangan pada panjang/tinggi dan lebar dari image

```python
def modus(image):
    arr = np.sort (image)
    indeks = np.zeros ( np.max(arr)+1, dtype= int )
    for i in range (len(arr)):
        indeks[arr[i]] += 1
        
    m = np.max(indeks)
    for j in range (len(indeks)):
        if (indeks[j] == m):
            return j
def modusFilter(image, h, w):
    panjang , lebar = image.shape
    result = np.zeros(image.shape, dtype=int)
    h2, w2 = int ((h-1)/2), int ((w-1)/2)
    pad = np.pad (image, ((h2,h-h2-1),(w2,w-w2-1)) , 'edge' )
    for i in range (panjang):
        for j in range (lebar):
            result[i,j] = modus(pad[i:i+h, j:j+w])
    return result
```

Kode diatas merupakan kode yag digunakan untuk menentukan modus filtering. Langkah yang dilakukan yaitu dengan memasukkan nilai matriksnya terlebih dahulu kemudian menghitung matriksnya dengan menghitung angka yang paling sering muncul kemudian dilakukan pengecekan angka yang sering muncul. Hasil yang diperoleh di olah lagi dengan cara melakukan perulangan pada tinggi dan lebar dari image

```python
plt.figure(figsize=(15,10))
plt.subplot(1,3,1)
plt.imshow(modusFilter(gray,3,3),cmap='gray')
plt.subplot(1,3,2)
plt.imshow(modusFilter(gray,5,5),cmap='gray')
plt.subplot(1,3,3)
plt.imshow(modusFilter(gray,7,7),cmap='gray')
```
