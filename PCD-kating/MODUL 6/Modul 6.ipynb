```python
import numpy as np
import PIL.Image as p
import matplotlib.pyplot as plt
```

```python
sx=np.array([[-1,0,1],
            [-2,0,2],
            [-1,0,1]])
sy=np.array([[1,2,1],
            [0,0,0],
            [-1,-2,-1]])
px=np.array([[-1,0,1],
            [-1,0,1],
            [-1,0,1]])
py=np.array([[1,1,1],
            [0,0,0],
            [-1,-1,-1]])
rx=np.array([[1,0],[0,-1]])
ry=np.array([[0,1],[-1,0]])
c=p.open("Modul 6.jpeg")
g=np.array(c)
plt.imshow(c)
```

```python
gr=c.convert('L')
plt.imshow(gr,cmap='gray')
a=np.array(gr)
X,Y=gr.size
print(X,Y)
```

```python
Sx=np.zeros(a.shape)
for i in range (1,Y-2):
    for j in range (1,X-2):
        Sx[i][j]=a[i-1,j-1]*sx[0,0] + a[i-1, j]*sx[0,1] + a[i-1,j+1]*sx[0,2] + a[i,j-1]*sx[1,0] + a[i, j]*sx[1,1] + a[i,j+1]*sx[1,2] + a[i+1,j-1]*sx[2,0] + a[i+1, j]*sx[2,1] + a[i+1,j+1]*sx[2,2]
plt.imshow(Sx, cmap='gray')
```

```python
Sy=np.zeros(a.shape)
for i in range (1,Y-2):
    for j in range (1,X-2):
        Sy[i][j]=a[i-1,j-1]*sy[0,0] + a[i-1, j]*sy[0,1] + a[i-1,j+1]*sy[0,2] + a[i,j-1]*sy[1,0] + a[i, j]*sy[1,1] + a[i,j+1]*sy[1,2] + a[i+1,j-1]*sy[2,0] + a[i+1, j]*sy[2,1] + a[i+1,j+1]*sy[2,2]
plt.imshow(Sy, cmap='gray')
```

```python
Ms=abs(Sx)+abs(Sy)
plt.imshow(Ms, cmap='gray')
```
