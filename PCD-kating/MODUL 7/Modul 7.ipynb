```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
c=cv2.imread('tes.jpg',0)
r,t=cv2.threshold(c,127,225,cv2.THRESH_BINARY)
k=np.ones((5,5),np.uint8)
```

```python
d=cv2.dilate(t,k,iterations=1)
plt.subplot(131),plt.imshow(c,cmap = 'gray')
plt.title('Citra Awal'), plt.xticks([]), plt.yticks([])
# plt.subplot(132),plt.imshow(thresh,cmap = 'gray')
# plt.title('Citra Biner'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(d,cmap = 'gray')
plt.title('Hasil Dilasi'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
e=cv2.erode(t,k,iterations=1)
plt.subplot(131),plt.imshow(c,cmap = 'gray')
plt.title('Citra Awal'), plt.xticks([]), plt.yticks([])
# plt.subplot(132),plt.imshow(thresh,cmap = 'gray')
# plt.title('Citra Biner'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(e,cmap = 'gray')
plt.title('Hasil Erosi'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
o=cv2.morphologyEx(t,cv2.MORPH_OPEN,k)
plt.subplot(131),plt.imshow(c,cmap = 'gray')
plt.title('Citra Awal'), plt.xticks([]), plt.yticks([])
# plt.subplot(132),plt.imshow(t,cmap = 'gray')
# plt.title('Citra Biner'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(o,cmap = 'gray')
plt.title('Hasil Opening'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
cl=cv2.morphologyEx(t,cv2.MORPH_CLOSE,k)
plt.subplot(131),plt.imshow(c,cmap = 'gray')
plt.title('Citra Awal'), plt.xticks([]), plt.yticks([])
# plt.subplot(132),plt.imshow(t,cmap = 'gray')
# plt.title('Citra Biner'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(cl,cmap = 'gray')
plt.title('Hasil Closing'), plt.xticks([]), plt.yticks([])
plt.show()
```
