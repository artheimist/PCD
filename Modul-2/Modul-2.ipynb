```python
img=image.imread('EmiliaForLife.jpg')
plt.imshow(img)
```

Code "img=image.imread('EmiliaForLife.jpg')" digunakan untuk membaca sebuah image dan "plt.imshow(img)" untuk menampilkan image yg sudah dibaca

```python
gray_image=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
```

Code "gray_image=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)" digunakan untuk merubah gambar yang sudah dibaca menjadi sebuah grayscale dan "plt.imshow(gray_image, cmap='gray')" untuk menampilkan gambar grayscale

```python
translasi=np.zeros(gray_image.shape)

for x in range (256):
    for y  in range (256):
        x2=x+50
        y2=y+50
        if x2<256:
            if y2<256: 
                translasi[x2][y2]=gray_image[x][y]
plt.imshow(translasi,cmap='gray')
```

code "translasi=np.zeros(gray_image.shape)" digunakan untuk membuat sebuah array kosong atau bernilai 0 dengan ukuran yang sama seperti ukuran image hasil grayscale. "for x in range (256)" dan "for y  in range (256):" digunakan untuk melakukan perulangan dengan range 256."x2=x+50"dan"y2=y+50" digunakan untuk menyatakan bahwa pergeseran sumbu x dan sumbu y masing-masing ditambahkan dengan 50. " if x2<256"dan"if y2<256:" merupakan sebuah kondisi yang dimana jika kondisi itu sesuai, maka akan melanjutkan ke proses selanjutnya yaitu "translasi[x2][y2]=gray_image[x][y]" yang dimana merupakan variabel hasil translasi dan menampilkan image dalam bentuk grayscale

```python
p = len(gray_image)
flip=np.zeros(gray_image.shape)
m = p-1 
for i in range (p): 
    for j in range (p): 
        flip[i][j] = gray_image [m][j] 
    m-=1
plt.imshow(flip,cmap='gray')
```

```python
pq = len(gray_image)
flip1=np.zeros(gray_image.shape)
m = pq-1 
for i in range (pq): 
    for j in range (pq): 
        flip1[i][j] = gray_image [j][m] 
    m-=1
plt.imshow(flip1,cmap='gray')
```

"flip1[i][j] = gray_image [j][m] " merupakan perhitungan untuk menghasilkan pencerminan pada sumbu y."m-=1" digunakan untuk mengurangi setiap nilai dari "m" dengan 1 dan dan akan berlanjut sesuai dengan panjang pixel

```python
x, y = gray_image.shape
def rotasi(degree):
    rotation = np.zeros((x, y))
    angle = -degree*(np.pi/180)

    center_x = x // 2
    center_y = x // 2

    for i in range(x):
        for j in range(y):
            x2 = int((i - center_x) * np.cos(angle) - (j - center_y) * np.sin(angle) + center_x)
            y2 = int((i - center_x) * np.sin(angle) + (j - center_y) * np.cos(angle) + center_x)
            if 0 <= x2 < x and 0 <= y2 < y:
                rotation[i][j] = gray_image[x2][y2]
    plt.imshow(rotation, cmap="gray")

rotasi(30)
```
