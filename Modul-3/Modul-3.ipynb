```python
import numpy as np #untuk memanggil library numpy dialiaskan sebagai np
import cv2  #untuk memanggil library cv2
import matplotlib.image as mpimg  #untuk memanggil library matplotlib
import matplotlib.pyplot as plt #untuk memanggil library matplotlib.pyplot dialiaskan sebagai plt
#Import Library
```

```python
img = mpimg.imread("EmiliaForLife.jpg") #untuk membaca image menggunakan library cv2
plt.imshow(img)  #untuk menampilkan gambar dengan parameter img
```

```python
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)  #untuk mengkonversi image menjadi grayscale image
plt.imshow(gray_img,cmap='gray') #untuk menampilkan hasil grayscale image dengan color map gray
```

```python
matriks=gray_img
print(gray_img) 
```

```python
k=np.zeros(256)
dk=np.array([i for i in range (256)])
for i in matriks:
    for j in i:
        k[j]+=1 
#melakukan suatu perulangan untuk mencari jumlah atau banyaknya masing" derajat keabuan dalam citra, perhitunggan ini dilakukan dengan cara  me increment (+1),disini juga dibuatkan sebuah array untuk menampung hasil perhitungan dari perulangan. 
```

```python
for x in range (256):
    k[x]/=65536
plt.xlabel('derajat keabuan')
plt.ylabel('probabilitas')
plt.bar(dk,k)

#untuk menghitung nilai prpbabilitas pada histogram, yang jumlah masing-masing derajat keabuan akan dibagi dengan ukuran dari citra tersebut yaitu 256 x 256 pixel, yang mana hasilnya 65536, setelah itu ditunjukkanlah hasil dari berbntuk diagram
```

```python
#JURNAL NO 2
c=np.zeros(dk.shape)
for x in range (len(dk)):
    for y in range (x+1):
        c[x]=c[x]+k[y]
c
#melakukan perhitungan untuk mencari nilai sk yaitu melalui perulangan dimana dilakukan dengan menjumlahkan nilai dr k dengan c, sehinggan diperoleh hasil c, perulangan terus berlanjut sampai batas yang ditentukan
```

```python
p = np.zeros(dk.shape)
p = np.round(c*255)
p = p.astype(int)
print (p)
sk = np.zeros(dk.shape)
for x in range (len(dk)):
    sk [ p[x]] += k[x]
sk

#potongan code diatas untuk mencari nilai frekuensi kumulutatif, yang mana dilakukan dengan membuat array kosong untuk menampung nilai dari p, setelah itu dilakukan pembulatan dari hasil kali c dengan 255, 22 ini di dapat dari pangjang matriks dikurangi 1,  setelah itu dibuat kedalam tipe data int hal ini dilakukan supaya angka dibelakang koma tidak diikut sertakan, setelah itu dilakukan suatu perulangan untuk mencari sknya
```

```python
plt.xlabel ('keabuan')
plt.ylabel ('prob keabuan')
plt.bar (dk, sk)

#kode tersebut untuk menampilkan histogram hasil ekualisasi
```

```python
for i in range (256):
    for j in range (256):
        matriks[i][j] = p[matriks[i][j]]

# insert hasil perhitungan ekualisasi ke dalam matriks gray img
```

```python
plt.imshow(matriks,cmap="gray")
```

```python
image = cv2.imread("subaru.jpg") #berfungsi untuk memanggil gambar yang terdapat didalam file 
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY) #berfungsi sebagai mengconvert color dari yang 3 layer menjadi satu layer yaitu abu saja, yang sbelumnya ada blue green dan red dan digabung menjadi satu layer yang menghasilkan warna abu
plt.imshow(gray,cmap="gray") #untuk merubah warnanya menjadi gray atau abu abu
```

```python
matriks = gray #deklarasi gray akan dibaca matriks
x = [i for i in range (256)] #perulangan untuk pemanggilan range 256
a = np.zeros(256) #untuk membuat 1 array 0 sebnyak 256 angka yang mana dalam angkanya terdapat 255 angka
for row in matriks: #merupakan perulangan yang terdapat pada baris pada nilai ukuran matriks pada gambar
    for col in row: #merupakan perulangan yang terdapat pada kolom pada nilai ukuran baris
        a[col]+=1 #setiap perulangan akan bertambah 1 setiap perulangan yang telah dijalankan
a = a/65536 #lalu dibagi dengan 65536 
plt.bar(x,a) #unutk menampilkan histogram nya
```
